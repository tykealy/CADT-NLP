**Removing Punctuation**

Exercise 6
+ Use the maketrans() function to define a translation table which maps each punctuation sign to an empty string. Use the maketrans function with 3 arguments where the third argument is the list of punctuation signs.
+ Define a function (called remove_punctuation) which segments, tokenizes and removes punctuation signs
 + Use the segment_and_tokenize function your defined in Exercise 5
 + Apply the translation table you just defined to each character of each token
+ Apply this function to the file
+ Print out the result

```python
# define a translation table here
import string
def remove_punctuation(input):
  remove_punc = str.maketrans('','',string.punctuation)
  words = segment_and_tokenize(input)
  return [[word.translate(remove_punc) for word in sentence] for sentence in words]

remove_punctuation(text)
  #define your function here
```

**Lowercasing**

Exercise 7:

+ Define a function (called lower_case) which segments, tokenizes, removes punctuation and lowercases all tokens
 + Use the remove_punctuation function you defined in Exercise 6 then
Lower case each character of each token
+ Apply this function to the file
+ Print out the result

```python
#define your function here
def lower_case(text):
  words = remove_punctuation(text)
  return [[word.lower() for word in sentence] for sentence in words]

lower_case(text)
```

**Remove Stop Words**


Exercise 8:
+ Define a function which segments, tokenizes, removes punctuation, lowercases all tokens and removes stop words
  + Use the lower_case functiond you defined in Exercise 7
  + Use this list to remove all stop words from the tokenized sentences
    + Hint: Python: the syntax "w not in l" allows you to check that string w is not in the list l

+ Apply this function to the file
+ Print the result

```python
from nltk.corpus import stopwords

def preprocessing(text):
   words = lower_case(text)
   en_stopwords = stopwords.words('english')
   return [[word for word in sentence if word not in en_stopwords] for sentence in words]

preprocessing(text)
   #define your function here
```

**Lemmatizing**

Exercise 9:
Now we would like to collect all verbs that appear in the text. The same verb occurs in different forms however (e.g., 'is, was, were'). To gather together these forms into a single form ('be'), we can use NLTK WordNet lemmatizer (nltk.stem.wordnet.WordNetLemmatizer) which given a token and its POS tag will return its lemma i.e., the word form which is usually found in dictionary entries.

```python

sentence = "Amélie is a story about a girl named Amélie whose childhood was suppressed by her Father's mistaken concerns of a heart defect."

# Tokenize and Pos-TAGs the text
tokens = preprocessing(sentence)
tagged_tokens = nltk.pos_tag(tokens[0])
tagged_tokens
# Specify which POS tags label verbs
tags = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
# Collect all verb tokens
verbs = [[tokens for tokens, tag in tagged_tokens if tag in tags]]
# Instantiate a lemmatizer object
lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()
# Lemmatize the verbs and collect the lemmas into a list
lemmas = [lemmatizer.lemmatize(verb, 'v') for verb in verbs[0]]
lemmas
```
